### Theoritical Probability


```python
def coin_flip_probabilty():
    heads_probability = 1/2
    return heads_probability

def rolling_dice_prob(number):
    if number not in range(1,7):
        raise ValueError("Number must be in between 1 to 6")
    probability = 1/6
    return probability   
```

```python
## Example Usage

coin_flip_result = coin_flip_probabilty()
dice_roll_result = rolling_dice_prob(4)

print("Probability of head in a coin flip : ", coin_flip_result)
print("Probability of rolling a 4 on a die :", dice_roll_result)
```

```python
### Experimental 1

import random
import pandas as pd
```

```python
def simulate_coin_flips(num_flips):
    results = []
    for _ in range(num_flips):
        flip = random.choice(['Heads' , 'Tails'])
        results.append(flip)
    return results
```

```python
pd.Series(simulate_coin_flips(100)).value_counts()
```

```python
def calculate_experimental_probabilty(results , event):
    event_count = results.count(event)
    total_count = len(results)
    probaility = event_count / total_count
    return probaility
```

```python
## Example Usage
num_flips = 1000
simulation_result = simulate_coin_flips(num_flips)
heads_probabilty = calculate_experimental_probabilty(simulation_result , "Heads")
tails_probabilty = calculate_experimental_probabilty(simulation_result , "Tails")

print("Number of flips " , num_flips)
print("Experimental Probability of heads : " , heads_probabilty)
print("Experimental Probability of tails : " , tails_probabilty)
```

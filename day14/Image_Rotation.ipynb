#### Rotation using matrix

```python
#load the image
image = cv2.imread('images/lenna.png')

#Define the centre point and rotation angle
center = (image.shape[1] // 2, image.shape[0] // 2)
angle = 60

#Create the rotation matrix 
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

#apply the rotation using wrapAffline
rotated_image = cv2.warpAffine(image, rotation_matrix, 
                                (image.shape[1], image.shape[0]))

#show the rotated image
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# using a different point in image as centre of rotation
#load the image
image = cv2.imread('images/lenna.png')

#Define the centre point and rotation angle
#center = (image.shape[1] // 2, image.shape[0] // 2)
angle = 60

#Create the rotation matrix 
rotation_matrix = cv2.getRotationMatrix2D((100, 255), angle, 1.0)

#apply the rotation using wrapAffline
rotated_image = cv2.warpAffine(image, rotation_matrix, 
                                (image.shape[1], image.shape[0]))

#show the rotated image
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# using a different point in image as centre of rotation
#load the image
image = cv2.imread('images/lenna.png')

#Define the centre point and rotation angle
#center = (image.shape[1] // 2, image.shape[0] // 2)
angle = 60

#Create the rotation matrix 
rotation_matrix = cv2.getRotationMatrix2D((100, 255), angle, 1.0)

#resizing
#apply the rotation using wrapAffline
rotated_image = cv2.warpAffine(image, rotation_matrix, 
                                (700, 700))

#show the rotated image
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

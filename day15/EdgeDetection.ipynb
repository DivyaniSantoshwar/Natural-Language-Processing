```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("images/lenna.png", 0)
```

```python
#sobel parameters :ksize,dx,dy,ddepth
sobelx = cv2.Sobel(image,cv2.CV_64F ,1, 0, ksize=3) # sobel for horizontal edges 
sobely = cv2.Sobel(image,cv2.CV_64F ,0, 1, ksize=3) # sobel for vertical edges 

#Combine the Sobel img (optional)
combined_sobel =  cv2.add(sobelx,sobely)
```

```python
thresh, binary_sobel = cv2.threshold(combined_sobel,120,255,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,5,1) 
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.subplot(1,5,2)
plt.title("Sobel X")
plt.imshow(sobelx, cmap="gray") 
plt.subplot(1,5,3) 
plt.title("Sobel y")
plt.imshow(sobely, cmap='gray')
plt.subplot(1,5,4) 
plt.title("Final ")
plt.imshow(combined_sobel, cmap='gray')
plt.subplot(1,5,5) 
plt.title("Final ")
plt.imshow(binary_sobel, cmap='gray')
```

```python
img1 = cv2.imread('images/circle.png',0)
```

```python
sobelx = cv2.Sobel(img1,cv2.CV_64F ,1, 0, ksize=3) # sobel for horizontal edges 
sobely = cv2.Sobel(img1,cv2.CV_64F ,0, 1, ksize=3) # sobel for vertical edges

#Combine the Sobel img (optional)
combined_sobel =  cv2.add(sobelx,sobely)
thresh, binary_sobel = cv2.threshold(combined_sobel,0,255,cv2.THRESH_BINARY)
```

```python
#Apply canny edge detection
edge = cv2.Canny(img1,50,150)

plt.figure(figsize=(16,8))
plt.subplot(1,3,1) 
plt.title("Original Image")
plt.imshow(img1, cmap="gray")
plt.subplot(1,3,2)
plt.title("Sobel")
plt.imshow(binary_sobel, cmap="gray") 
plt.subplot(1,3,3) 
plt.title("Canny ")
plt.imshow(edge, cmap='gray')
```

```python
#load the images
gray = cv2.imread('images/circle.png',0)
image= cv2.imread('images/circle.png')

#convert image to float32
gray = np.float32(gray)

#apply the harris corner detector
dst = cv2.cornerHarris(gray,blockSize=2,ksize=3,k=0.04)

#result is dilated for marking the corner not important
dst = cv2.dilate(dst,None)

#Threshold for an optimal value , it may vary depending on the image
image[dst>0.01*dst.max()]=[0,0,255]

#Display the result
cv2.imshow('Harris Corner',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

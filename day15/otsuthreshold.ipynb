```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the images 
gray =cv2.imread("images/manas.jpeg", 0  )
image =cv2.imread("images/manas.jpeg",0)

#Apply Otsu's Thresholding
value, thresh_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU)
```

```python
#Display results
ret, thresh1 = cv2.threshold(image, 120,255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(image, 120,255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(16,8))
plt.subplot(1,3,1) 
plt.title("Original Image")
plt.imshow(gray, cmap="gray")
plt.subplot(1,3,2)
plt.title("Binary Image")
plt.imshow(thresh1, cmap="gray") 
plt.subplot(1,3,3) 
plt.title("Otsu Image")
plt.imshow(thresh2, cmap='gray')
```

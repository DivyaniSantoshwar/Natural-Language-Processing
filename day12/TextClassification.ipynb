#### Data Preparation

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
sent = 'Hello friends ! How are you? I like Python Programming.' 
```

```python
#1. Tokenize the text
tokens1 = word_tokenize(sent)
tokens1
```

```python
#2. Remove the punctuation
tokens2 = [token for token in tokens1 if token.isalnum()]
tokens2
```

```python
#3. Remove the stopwords
swords = stopwords.words('english')
tokens3 = [token for token in tokens2 if token.lower() not in swords]
tokens3
```

```python
#4. Remove the suffixes from words
ps = PorterStemmer()
```

```python
for word in tokens3:
    print(word , ':' , ps.stem(word))
```

```python
tokens4  = [ps.stem(token) for token in tokens3]
tokens4
```

```python
def clean_text(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalnum()]
    tokens3 = [token for token in tokens2 if token.lower() not in swords]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4
```

```python
clean_text(sent)
```

#### TF*IDF VECTORIZER

```python
#Import the class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#create the object
tfidf = TfidfVectorizer(analyzer=clean_text)
```

```python
#apply
x_new = tfidf.fit_transform(x)
x_new
```

#### Cross-Validation

```python
#import the function

from sklearn.model_selection import train_test_split
```

```python
#separate the training and testing data
x_train, x_test, y_train, y_test = train_test_split(x_new, y, random_state=0)
x_train.shape
```

```python
x_test.shape
```

```python
y_train
```

```python
y_test
```

```python
tfidf.get_feature_names_out()
```

#### Build the machine learning classification model

```python
#import the class
from sklearn.svm import SVC
```

```python
#create the object
svm = SVC(kernel='sigmoid',random_state=0)
```

```python
%%time
svm.fit(x_train, y_train)
```

#### Performance Evaluation

```python
from sklearn.metrics import confusion_matrix , accuracy_score, ConfusionMatrixDisplay,classification_report
```

```python
y_pred = svm.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python

print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

### GUI for spamclassifier 

```python
import joblib
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from tkinter import *
import nltk
swords = stopwords.words('english')
ps = PorterStemmer()

def clean_text(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalnum()]
    tokens3 = [token for token in tokens2 if token.lower() not in swords]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4

top = Tk()
top.title('My first application')
top.geometry('800x600')

def check():
    classifier=joblib.load('classifier.model')
    tfidf= joblib.load('preprocessor.model')
    message = str(t.get("1.0", END)) 
    result = classifier.predict(tfidf.transform([message]))

    l.config(text='Message is :' + str(result[0]))
    l.place(x=350 , y=550)
    

Label(top, text='Enter your msg here:').place(x=50, y=30)
t = Text(top)
t.place(x=50, y=70)
Button(top, text="Check", command= check).place(x=350, y=500)
l=Label(top, text='Message is: ')

top.mainloop()
```
